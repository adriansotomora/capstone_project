## Clustering

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Select relevant behavioral features (you can adjust as needed)
features_for_clustering = [
    'total_volume', 'total_income', 'total_profit',
    'median_ticket', 'median_ticket_std',
    'median_promotor_calls', 'total_promotor_calls',
    'profit_per_visit', 'visit_order_gap', 'total_promotor_visits', 'avg_orders_per_month', 'efficiency'
]

# Step 2: Prepare and scale the data
X = clients[features_for_clustering].copy().fillna(0)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 3: Apply KMeans
k = 5  # Choose number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
clients['cluster'] = kmeans.fit_predict(X_scaled)

# Step 4: Visualize clusters with PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
clients['pca_1'] = X_pca[:, 0]
clients['pca_2'] = X_pca[:, 1]

plt.figure(figsize=(10, 6))
sns.scatterplot(data=clients, x='pca_1', y='pca_2', hue='cluster', palette='Set2')
plt.title('Client Clusters (PCA Projection)')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
# Summary of clusters
cluster_summary = clients.groupby('cluster')[features_for_clustering].mean().round(2)
display(cluster_summary)
```

```python
# Assuming 'class' and 'cluster' columns exist in `clients`
cluster_class_counts = clients.groupby(['class', 'cluster']).size().unstack(fill_value=0)

# Display the count table
print(cluster_class_counts)
```

```python
#average efficiency per cluster

efficiency_per_cluster = clients.groupby('cluster')['efficiency'].mean().round(2)
efficiency_per_cluster
```

## Model

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from xgboost import XGBRegressor, plot_importance
import matplotlib.pyplot as plt

# Define features and target
target = 'avg_visits_per_month'
features = [
    "total_volume", "total_income", "total_profit", "median_ticket", 
    "median_ticket_std", "median_promotor_calls", "total_promotor_calls", 
    "profit_per_visit", "visit_order_gap"
]

# Store predictions per cluster
all_cluster_predictions = []

# Iterate over each cluster
for cluster in clients['cluster'].unique():
    cluster_data = clients[clients['cluster'] == cluster]

    # Split efficient and inefficient clients
    efficient_data = cluster_data[cluster_data['efficiency'] >= 1]
    inefficient_data = cluster_data[cluster_data['efficiency'] < 1]

    # Skip cluster if insufficient data
    if len(efficient_data) < 10 or len(inefficient_data) == 0:
        print(f"❌ Skipping cluster {cluster}: Not enough data")
        continue

    # Prepare features
    X_train = efficient_data[features].fillna(0)
    y_train = efficient_data[target]
    X_test = inefficient_data[features].fillna(0)

    # Scale features
    scaler = StandardScaler()
    X_train_scaled = pd.DataFrame(scaler.fit_transform(X_train), columns=features, index=X_train.index)
    X_test_scaled = pd.DataFrame(scaler.transform(X_test), columns=features, index=X_test.index)

    # Train model
    model = XGBRegressor(objective="reg:squarederror", random_state=42)
    model.fit(X_train_scaled, y_train)

    # Evaluate only on training data (efficient clients)
    y_train_pred = model.predict(X_train_scaled)
    rmse = np.sqrt(mean_squared_error(y_train, y_train_pred))
    r2 = r2_score(y_train, y_train_pred)
    print(f"\n📊 Cluster {cluster} — RMSE (Efficient): {rmse:.2f} | R²: {r2:.2f}")

    # Show feature importance
    plt.figure(figsize=(8, 5))
    plot_importance(model, importance_type='gain', title=f'Feature Importance — Cluster {cluster}')
    plt.tight_layout()
    plt.show()

    # Predict for inefficient clients only
    y_pred = model.predict(X_test_scaled)
    inefficient_data = inefficient_data.copy()
    inefficient_data["predicted_avg_visits_per_month"] = y_pred
    all_cluster_predictions.append(inefficient_data)

# Combine predictions for all inefficient clients
final_predictions_df = pd.concat(all_cluster_predictions)

# Compute and show average actual vs predicted visits per cluster
avg_actual = final_predictions_df.groupby("cluster")["avg_visits_per_month"].mean().round(2)
avg_predicted = final_predictions_df.groupby("cluster")["predicted_avg_visits_per_month"].mean().round(2)

avg_predicted_visits = pd.DataFrame({
    "avg_visits_per_month (inefficient)": avg_actual,
    "predicted_avg_visits_per_month": avg_predicted
}).reset_index()

print("\n🔍 Comparison of actual vs. predicted visits (inefficient clients only):")
print(avg_predicted_visits)
# Sample output
print(final_predictions_df[['client_id', 'cluster', 'avg_visits_per_month', 'predicted_avg_visits_per_month']].head())
```

```python
# 📊 Average actual and predicted visits per cluster (only for inefficient clients)

# Group by cluster and compute mean values
avg_visits_per_cluster = final_predictions_df.groupby('cluster')['avg_visits_per_month'].mean()
predicted_visits_per_cluster = final_predictions_df.groupby('cluster')['predicted_avg_visits_per_month'].mean()

# Calculate percentage change: (predicted - actual) / actual * 100
percentage_change = ((predicted_visits_per_cluster - avg_visits_per_cluster) / avg_visits_per_cluster * 100).round(2)

# Combine results into a single DataFrame
avg_predicted_visits = pd.DataFrame({
    'avg_visits_per_month (inefficient)': avg_visits_per_cluster.round(2),
    'predicted_avg_visits_per_month': predicted_visits_per_cluster.round(2),
    'percentage_change (%)': percentage_change
}).reset_index()

# Display results
print("🔍 Comparison of actual vs. predicted visits (inefficient clients only):")
print(avg_predicted_visits)
```

```python
# sum of saved visits

total_saved_visits = ((final_predictions_df['avg_visits_per_month'] - final_predictions_df['predicted_avg_visits_per_month']).sum()) * 12

total_saved_visits
```

```python
# saved cost of saved visits

total_saved_cost = total_saved_visits * 15

total_saved_cost
```
